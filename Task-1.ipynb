```python
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
for i in marks:
    if i>=50:
        pass_count+=1
    else:
        fail_count+=1
print("Total Pass Count:",pass_count)
print("Total Fail Count:",fail_count)
```

```python
names=[" Alice ","bob"," CHARLIE "]
cleaned_names=[]
for name in names:
    cleaned_name=name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Cleaned names:",cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    print("Message:",msg)
    print("Length:",length)
    if length>10:
        print("This message is more than 10 characters longer")
    print()
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"] 
err_count=0
for l in logs:
    if l=="ERROR":
        err_count+=1
print("The Error Count is :", err_count)
```
