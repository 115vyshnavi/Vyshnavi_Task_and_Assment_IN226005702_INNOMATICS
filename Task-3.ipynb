```python
query="Buy mobile phone by phone online"
#lowercase
query=query.lower()
#spliting
words=query.split()
result={}
for word in words:
   if words.count(word)>1:
       result[word]=words.count(word)
print(result)
```

```python
sensor_reading=[3,4,7,8,10,12,5]
valid_reading=[]
for i in range(len(sensor_reading)):
    #checking even 
    if sensor_reading[i]%2==0:
        valid_reading.append((i, sensor_reading[i]))
print("Valid sensor readings (Hour, Value):")
print(valid_reading)
```

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count={}
for email in emails:
    domain=email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] +=1
    else:
        domain_count[domain]=1
#total mails
total=len(emails)
for domain in domain_count:
    percent=(domain_count[domain]/total)*100
    print(f"{domain}:{int(percent)}%")
```

```python
sales=[1200, 1500, 900, 2200, 1400, 300]
avg=sum(sales)/len(sales)
for i in range(len(sales)):
    if sales[i]>avg * 1.3:
        print(f"Day {i+1}: {sales[i]}")
```
