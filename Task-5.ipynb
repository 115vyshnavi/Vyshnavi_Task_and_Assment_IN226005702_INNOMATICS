```python
def rainfall_checker(rainfalls, need):
    #Calculte avg rainfall
    total=0
    for rain in rainfalls:
        total = total + rain
    avg=total/len(rainfalls)
    #chcek if rainfall is adequate
    if avg>=need:
        status="Adequate Rainfall"
    else:
        status="Inadequate Rainall"
    #print the result
    print("Average Rainfall:",int(avg))
    print("Rainfall Status:",status)
#usage
rainfall_data=[60,80,70,78]
required_level=70
rainfall_checker(rainfall_data, required_level)
```

```python
def check_duplicate_acc(usernames):
    unique_users=set()
    for user in usernames:
        if user in unique_users:
            print("Dupicate AccountsFound: Yes")
            return 
        unique_users.add(user)
    print("Duplicate Acounts Found: No")
usernames=["user1", "user2", "user3", "user1"]
check_duplicate_acc(usernames)
```

```python
def premium_crop_filter(prices):
    premium_crops=[price for price in prices if price>2000]
    return premium_crops
crop_prices=[1500,2500,1800,3200]
premium_prices=premium_crop_filter(crop_prices)
print("Premium Crops:", premium_prices)
```

```python
def check_eligibility(total,attended):
    percent=(attended/total)*100
    if percent >=75:
        return percent, "Eligible"
    else:
        return percent, "Not Eligible"
total=100
attended=80
percent,status=check_eligibility(total, attended)
print("Attendance Percentage:", round(percent, 1))
print("Exam Eligibility:",status)
```
