```python
def manage_parking(capacity, logs):
    current_parked=0
    peak_usage=0
    for action in logs:
        if action=="IN":
            current_parked +=1
        elif action == "OUT":
            current_parked -=1
        #tack highest no.of cars 
        if current_parked > peak_usage:
            peak_usage=current_parked
    if current_parked >= capacity:
        status="FULL"
    else:
        status="Available"
    print("Current Parked Vechile:",current_parked)
    print("Parking Status:",status)
manage_parking(5,["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
def estimate_delivery(distance, traffic, weather):
    base_time=distance*5
    delay=0
    if traffic=="High":
        delay +=20
    elif traffic =="Medium":
        delay +=10
    if weather =="Rainy":
        delay +=15
    elif weather =="Snowy":
        delay ==25
    total_eta= base_time+delay
    print("Estimated Deliver Time:", total_eta,"minutes")
estimate_delivery(8, "High", "Rainy")
```

```python
def analyze_occupancy(total_seats, booked_seats_list):
    booked_count = len(booked_seats_list)
    occupancy_pct = (booked_count / total_seats) * 100
    if occupancy_pct == 100:
        status = "Housefull"
    elif occupancy_pct >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"
    print(f"Occupancy: {int(occupancy_pct)}%")
    print(f"Show Status: {status}")
booked_data = [1] * 150 
analyze_occupancy(200, booked_data)
```

```python
def classify_server_load(cpu_readings):
    average_cpu = sum(cpu_readings) / len(cpu_readings)
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"
    print("Average CPU Load:", average_cpu,"%")
    print("Server Status:",status)
classify_server_load([45, 60, 70, 85, 90])
```

```python
def monitor_resources(usage_dict):
    overused = []
    threshold = 7 # Example threshold
    for resource, hours in usage_dict.items():
        if hours > threshold:
            overused.append(resource)
    energy_alert = "Yes" if len(overused) > 0 else "No"
    print(f"Overused Resources: {', '.join(overused)}")
    print("Energy Alert:", energy_alert)
usage = {"Projector": 6, "AC": 9, "Lights": 4}
monitor_resources(usage)
```

```python
def manage_registrations(capacity, total_regs):
    confirmed = 0
    waitlist = 0
    if total_regs > capacity:
        confirmed = capacity
        waitlist = total_regs - capacity
        status = "Closed"
    else:
        confirmed = total_regs
        waitlist = 0
        status = "Open"
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlist}")
    print(f"Registration Status: {status}")
manage_registrations(100, 105)
```
