```python
class SmartLight:
    def __init__(self, name):
        self.name=name
    def set_status(self, status):
        print(self.name, "is", status)
name="Bedroom Light"
action="ON"
Light=SmartLight(name)
Light.set_status(action)
```

```python
class Employee:
    def __init__(self, name, e_id, dept):
        self.name = name
        self.id = e_id
        self.dept = dept
    def display_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.id)
        print("Department:", self.dept)
emp = Employee("Rahul", "EMP102", "IT")
emp.display_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name=name
        self.phone=phone
    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
c=Contact("Anitha", "9876543210")
c.show_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name=name
        self.price=price
    def print_tag(self):
        print("Product:", self.name)
        print("Price: ₹",self.price)
p=Product("HeadPhones", 2499)
p.print_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name=name
        self.rating=rating
    def show_details(self):
        print("Movie:", self.name)
        print("Rating:",self.rating,"/5")
m=Movie("Inception",4.8)
m.show_details()
```

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer=customer
        self.address=address
    def display_details(self):
        print("Delivery Detals")
        print("Customer:", self.customer)
        print("Address:", self.address)
d=Delivery("Suman", "Hyderabad")
d.display_details()
```
