```python
employees = {"Ravi": 75000, "Anita": 68000, "Kiran": 72000}
h_name=" "
highest_salary=0
for name in employees:
    if employees[name]>highest_salary:
        highest_salary=employees[name]
        h_name=name
print("Highest Salary:",h_name,"-",highest_salary)


```

```python
numbers=[45,22,89,10,66]
max_val=numbers[0]
min_val=numbers[0]
for num in numbers:
    if num>max_val:
        max_val=num
    if num<min_val:
        min_val=num
print("List:",numbers)
print("Max:",max_val)
print("Min:",min_val)
```

```python
prices=[450,1200,899,1500,300]
count=0
for p in prices:
    if p>1000:
        count +=1
print("Products above 1000:",count)
```

```python
attendance = ["P", "P", "A", "P", "P"]
present_count=0
for day in attendance:
    if day == "P":
        present_count += 1
percentage=(present_count/len(attendance))*100
print("Attendence Percentage:",percentage)
```

```python
text="python"
char_count={}
for ch in text:
    if ch in char_count:
        char_count[ch]+=1
    else:
        char_count[ch]=1
print(char_count)
```
